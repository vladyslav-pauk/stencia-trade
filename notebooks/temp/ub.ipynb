```python
import numpy as np

def make_point_clouds(n_samples, n_points, noise_level, n_dimensions=3):
    """
    Generate random point clouds with optional labels.

    Parameters:
    - n_samples (int): Number of point clouds to generate.
    - n_points (int): Number of points in each point cloud.
    - noise_level (float): Standard deviation of Gaussian noise to add to the points.
    - n_dimensions (int): Number of dimensions of each point cloud (default is 3).

    Returns:
    - point_clouds (ndarray): Array of shape (n_samples, n_points, n_dimensions) containing the point clouds.
    - labels (ndarray): Array of shape (n_samples,) containing labels (optional, set to zeros here).
    """
    point_clouds = []
    labels = []

    for i in range(n_samples):
        # Generate random points in the specified dimension
        points = np.random.rand(n_points, n_dimensions)
        # Add Gaussian noise
        points += np.random.normal(0, noise_level, points.shape)
        point_clouds.append(points)
        labels.append(i)  # Assign a label (e.g., unique ID for each point cloud)

    return np.array(point_clouds), np.array(labels)


# Usage Example
n_samples_per_class = 10
n_points = 500
noise_level = 0.1
n_dimensions = 40

point_clouds, labels = make_point_clouds(n_samples_per_class, n_points, noise_level, n_dimensions)

# Print the shape of the generated point clouds
print(f"There are {point_clouds.shape[0]} point clouds in {point_clouds.shape[2]} dimensions, "
      f"each with {point_clouds.shape[1]} points.")

# Plot one of the point clouds
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(point_clouds[0][:, 0], point_clouds[0][:, 1], point_clouds[0][:, 2], c='b', label="Point Cloud 0")
ax.set_title("3D Point Cloud Example")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()
plt.show()
```

```python
def plot_in_axis(ax, point_cloud):
    xs, ys, zs = zip(*point_cloud)
    ax.scatter(list(xs),list(ys),list(zs))
    return ax
```

```python
def plot_3d_data(point_cloud):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax = plot_in_axis(ax)
    plt.show()
```

```python
def plot_3_figures(point_cloud_1, point_cloud_2, point_cloud_3): 
    # Set up a figure three times as wide as it is tall
    fig = plt.figure(figsize=plt.figaspect(0.3))
    point_clouds = [point_cloud_1, point_cloud_2, point_cloud_3]
    for i in range(len(point_clouds)):
        axis = fig.add_subplot(1, 3, i+1, projection='3d')
        plot_in_axis(axis, point_clouds[i])
    plt.show()
    
circumference = point_clouds[0]
sphere = point_clouds[n_samples_per_class - 1]
torus = point_clouds[n_samples_per_class - 1]
plot_3_figures(circumference, sphere, torus)
```
